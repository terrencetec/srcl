```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal

import kontrol
import vishack


channel = "L1:LSC-SRCL_IN1_DQ"

data = vishack.data.diaggui.Diaggui("data/241113-SRCL_IN_OUT.xml")
f, asd = data.psd(channel)

ydata = asd[f>9]
xdata = f[f>9]

tf_fit = kontrol.curvefit.spectrum_fit(xdata, ydata, nzero=0, npole=0, clean=False)
```

```python
plt.loglog(f, asd, label=channel)
plt.loglog(f, abs(tf_fit(1j*2*np.pi*f)), label="Fit")
plt.legend(loc=0)
plt.grid(which="both")
plt.ylabel("Amplitude spectral density")
plt.xlabel("Frequency (Hz)")
```

```python
z, p, k = scipy.signal.tf2zpk(tf_fit.num[0][0], tf_fit.den[0][0])
np.savez("model_params/sens_noise.npz", zs=z, ps=p, k=k)
```
